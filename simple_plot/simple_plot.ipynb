```python
import matplotlib.pyplot as plt
import pandas as pd

# Create a DataFrame directly
data = {
    'model1': [0.53, 0.99, 0.22, 0.75, 0.45],
    'model2': [0.28, 0.65, 0.82, 0.55, 0.35],
    'model3': [0.07, 0.31, 0.44, 0.25, 0.15],
    'model4': [0.96, 0.66, 0.91, 0.85, 0.65]
}
df = pd.DataFrame(data, index=['task1', 'task2', 'task3', 'task4', 'task5'])

# Parameters for the plot
line_args = {"linewidth": 5, "alpha": 0.5}
scatter_args = {"s": 100, "alpha": 0.8}

# Create the plot
for col in df.columns:
    plt.plot(df.index, df[col], label=col, **line_args)
    plt.scatter(df.index, df[col], **scatter_args)

# Display the legend
plt.legend()

# Set y-axis limits
plt.ylim(0, 1)

# Show the plot
plt.show()
```
